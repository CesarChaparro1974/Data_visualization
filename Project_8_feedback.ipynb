**2. Visualizations: For this part I will work with seaborn library, to plot a stripplot visualization.**

```python
# first visualization from the DataFrame.
sns.stripplot(x="diet", y="weight", hue='sex', jitter=0.09, data=diet)
```

**I added the parameter jitter to add space between dots in order to give more clarity to the readibility of the plot.**

```python
# replace legend values 0 and 1 for Female and Male.
diet['sex'] = diet['sex'].replace({0:'Female', 1:'Male'})
sns.stripplot(x="diet", y="weight", hue='sex', jitter=0.09, data=diet)
```

```python
# replace x-axis values 1, 2, 3 for A, B, C.
diet['diet'] = diet['diet'].replace({1:'A', 2:'B', 3:'C'})
sns.stripplot(x="diet", y="weight", hue='sex', jitter=0.09, data=diet)
```

```python
# reorder the diet x-axis values as A, B, C.
diet_order = ['A', 'B', 'C']
sns.stripplot(x="diet", y="weight", hue='sex', jitter=0.09, data=diet, order=diet_order)
```

```python
# replace palette colors to match the result.
sns.stripplot(x="diet", y="weight", hue='sex', jitter=0.09, data=diet, order=diet_order, palette={'Male':'green', 'Female':'magenta'})
```

```python
# replace labels for x-axis and y-axis.
sns.stripplot(x="diet", y="weight", hue='sex', jitter=0.09, data=diet, order=diet_order, palette={'Male':'green', 'Female':'magenta'})
plt.xlabel('Diet Label')
plt.ylabel('Adult Weight/lbs')
```

```python
# add title to the plot.
sns.stripplot(x="diet", y="weight", hue='sex', jitter=0.09, data=diet, order=diet_order, palette={'Male':'green', 'Female':'magenta'})
plt.xlabel('Diet Label')
plt.ylabel('Adult Weight/lbs')
plt.title('Red Squirrel Diet Study')
```

```python
# resize the plot to increase its size.
plt.figure(figsize=(9, 7))
sns.stripplot(x="diet", y="weight", hue='sex', jitter=0.09, data=diet, order=diet_order, palette={'Male':'green', 'Female':'magenta'})
# resize font for x and y ticks for consistency on the visualization.
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
# resize font for x and y labels for consistency on the visualization.
plt.xlabel('Diet Label', fontsize=14)
plt.ylabel('Adult Weight/lbs', fontsize=14)
# resize font for title for consistency on the visualization.
plt.title('Red Squirrel Diet Study', fontsize=16)
```

**Description:<br>
The plot shows the weight of adult red squirrels based on their diet, categorized into three groups labeled as 'A', 'B', and 'C'.
The points represent individual squirrels, with their weights plotted along the y-axis.
Different colors (green and magenta) represent the sexes (Male and Female) of the squirrels.
Squirrels of both sexes experimented an increase of their weight more significally with Diet B, while with Diet A the increase was very insignificant, and being Diet C the type of diet and is con second place on the effectiveness to increase weight.**
